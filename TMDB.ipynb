Accessing Web Page from Internet

```python
def get_html(url) :
    try :
        source = requests.get(url)
        # If any issue arises ex. wrong url then below line will raise an error
        source.raise_for_status()
        return source.text
    except Exception as e:
        return print(e)

```

```python
html = get_html("https://www.imdb.com/chart/top/")

#print(html)
```

```python
# below line will raise an error because URL not exists
get_html("https://www.imdb.com/chart/topsssf/")
```

Extract HTML 

```python
soup = BeautifulSoup(html, 'html.parser') # other parsers are html5lib , lxml

#print(soup)
```

```python
# find will consider first occurrence   -- find(name, attrs, recursive, string)
# find_all will consider all occurences -- find_all(name, attrs, recursive, string, limit)
top_rated = soup.find('tbody', class_ = 'lister-list')

#print(top_rated)
```

```python
all_top_rated = top_rated.find_all('tr')
#single_from_all_top_rated = all_top_rated.find_all('td')

#print(single_from_all_top_rated)
```

```python
for top in all_top_rated :
    names = top.find('td', class_ = 'titleColumn').find('a')
    name_of_movie = names.text
    print(name_of_movie)
```
